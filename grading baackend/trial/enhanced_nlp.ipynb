```python
import spacy
nlp = spacy.load('en_core_web_trf')
```

```python
doc = nlp("I am Aayush Jha, and my brother is Piyush Jha. And My friend is Amit Jangir")
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
with open("Amit_jangir.txt", 'r') as f:
    Amit = f.read()
```

```python
amit_doc = nlp(Amit)
```

```python
for ent in amit_doc.ents:
    print(ent.text, ent.label_)
```
