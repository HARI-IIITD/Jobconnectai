```python
import re

text_data ='''Priya Gupta
Mechanical Engineer
Email: priya.gupta@gmail.com | Phone: +91 9123456780
Profile Summary
Mechanical Engineer with 5 years of experience in product design, CAD modeling, and
manufacturing systems.
Education
B.Tech in Mechanical Engineering, IIT Kanpur, 2014 - 2018
Skills
AutoCAD, SolidWorks, MATLAB, CNC Machining, Thermodynamics
Problem Solving Skills
Designed cost-efficient mechanical systems reducing manufacturing cost by 20%.
Work Experience
Mechanical Engineer, Larsen & Toubro, 2018 - Present, Worked on product design, FEA
simulations, and assembly line automation.
Projects
Smart Cooling System, Developed energy-efficient HVAC prototype.
Certifications
SolidWorks Professional, Six Sigma Green Belt
Awards & Recognitions
Best Innovator Award - L&T; (2021)
Hobbies
Painting, Robotics, Badminton'''

fields = ["Education", "Skills",
        "Experience", "Projects", "Certifications", "Hobbies", "Sector"
    ]

pattern = r"(?i)(" + "|".join(fields) + r")\s*\s*"
# for i in fields:
#       parts = re.split(i, text_data)
#       print(parts[0])
parts = re.split(pattern, text_data)
extracted = {}

for i in range(1, len(parts), 2):
        field = parts[i].strip().capitalize()
        content = parts[i+1].strip()
        # Stop content at the next field keyword if it's concatenated
        next_field_match = re.search(pattern, content)
        if next_field_match:
            content = content[:next_field_match.start()].strip()
        extracted[field] = content
    
print(extracted)
print(json.dumps(extracted, indent=2))
```
